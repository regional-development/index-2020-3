## p8_03
### кількість нещасних випадків на виробництві

```python
df_P08_006 = pd.read_excel(PATH_INTERIM / "P8" / "P08_006.xlsx")
df_P08_006["region"] = df_P08_006["region"].str.strip()
df_P08_006_fin = (
    df_P08_006
    .groupby("region")
    .agg(p8_03_raw = ("Код події", "size"))
    .reset_index()
)

df_P08_006_fin["p8_03"] = normalize_parameter(df_P08_006_fin["p8_03_raw"], reverse=True)
```

## p8_04
### частка неформально зайнятого населення

```python
df_P08_007 = pd.read_excel(PATH_INTERIM / "P8" / "P08_007.xlsx")
df_P08_007["region"] = df_P08_007["region"].str.strip()

df_P08_007["p8_04_raw"] = (
    df_P08_007["Кількість неформально зайнятих у віці 15-70 років у січні-червні 2020 року, тис. осіб"] / 
    df_P08_007["Робоча сила у віці 15-70 років у січні-червні 2020 року, тис. осіб"]
)

df_P08_007["p8_04"] = normalize_parameter(df_P08_007["p8_04_raw"], max_bound=1, reverse=True)
```
